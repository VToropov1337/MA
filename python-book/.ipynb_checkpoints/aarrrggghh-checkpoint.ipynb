```python
import pandas as pd
import numpy as np

df = pd.read_csv('raw_data.csv',dtype = {'bc': str, 'sku': str, 'producer': str, 'brand': str})
pd.set_option('display.max_columns', 100)
```

```python
brands = ['ЛЕНТА','365','ALPRO']
```

```python
df1 = df[(df['outlet'] == 'ЛЕНТА')& df['brand'].notnull()]
```

```python
df1 = df1.assign(is_stm = 0)
```

```python
for i in df1.index:
    for j in brands:
        if str(j) in str(df1.loc[i][['sku','brand','producer']]).upper():
            df1.loc[i,'is_stm'] = 1
```

```python
#store_title - название сети
#globus_title - массив строк-брендов
#рабочая ищет
def check_brand(df,store_title):
    df_store = df[df['outlet'] == str(store_title)]
    df_store = df_store[df_store['brand'].apply(lambda x: str(x) in brands)]
    return df_store
    
qq = check_brand(df1,'ЛЕНТА')
qq.head()
    
```
